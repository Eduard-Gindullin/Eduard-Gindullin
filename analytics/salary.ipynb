```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/ds_salaries.csv", index_col="Unnamed: 0")
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
df.info
```

Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)

```python
vacancy_count = len(np.unique(df['job_title']))
print(f'Количество уникальных должностей в датасете: {vacancy_count}')
```

```python
def categorize_job(list):
  if "AI" in list:
    return "AI"
  elif "Big Data" in list:
    return "BD"
  elif "Analy" in list:
    return "DA"
  elif "Research Scientist" in list:
    return "DA"
  elif "Data Engin" in list:
    return "DE"
  elif "ETL" in list:
    return "DE"
  elif "Data Architect" in list:
    return "DE"
  elif "Computer Vision" in list:
    return "DL"
  elif "NLP" in list:
    return "DL"    
  elif "Data Scien" in list:
    return "DS"
  elif "Head of Data" in list:
    return "DS"
  elif "Data Specialist" in list:
    return "DS"
  elif "Machine Learning" or "ML" in list:
    return "ML" 
  else:
    return "Other"
```

```python
df['job_category'] = df['job_title'].apply(categorize_job)
df
```

```python
df.loc[df['job_category'] == 'DS']['job_title'].value_counts()
```

Какая средняя и медианная зарплата по группам вакансий?

```python
np.round(pd.pivot_table(df, index=['job_category'], values=['salary_in_usd'], aggfunc=[np.mean, np.median]), 2)
```

Какая средняя и медианная зарплата по каждому региону?

```python
np.round(pd.pivot_table(df, index=['company_location'], values=['salary_in_usd'], aggfunc=[np.mean, np.median]), 2)
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
(df.groupby('job_title')['salary_in_usd'].mean()/1000).sort_values(ascending = False).head(10)
```

```python
(df.groupby('job_title')['salary_in_usd'].mean()/1000).sort_values(ascending = False).head(1)
```

Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
print(round(df.groupby('employee_residence')['salary'].count()/len(df), 5))
```
