```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = []
gender = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
with open("dataset_home.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(",")[0]))

    if human.split(",")[1] == 'male':
      gender.append(1)
    else:
      gender.append(2)
    
    bmi.append(float(human.split(",")[2]))

    children.append(int(human.split(",")[3]))

    if human.split(",")[4] == 'yes':
      smoker.append(1)
    else:
      smoker.append(2)

    if human.split(",")[5] == 'northwest':
      region.append(0)
    elif human.split(",")[5] == 'northeast':
      region.append(1)
    elif human.split(",")[5] == 'southwest':
      region.append(2)
    elif human.split(",")[5] == 'southeast':
      region.append(3)
    else:
      region.append(human.split(",")[5])

    charges.append(float(human.split(",")[6]))
```

```python
age = np.array(age)
gender = np.array(gender)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charges = np.array(charges)
```

```python
plt.scatter(np.log(charges[smoker>1]), age[smoker>1], label="кореляция")
plt.legend()
plt.show()
```

```python
plt.scatter(region, charges, label="типичные значения")
plt.legend()
plt.show()
```

```python
plt.scatter(children[age>20], charges[age>20], label="зависимость")
plt.legend()
plt.show()
```

```python
plt.scatter(bmi[age>50], charges[age>50], label="выбросы")
plt.legend()
plt.show()
```

```python
plt.scatter(np.log(charges[children>2]),bmi[children>2], label="Независимость")
plt.legend()
plt.show()
```

```python
plt.scatter(charges[smoker<2], age[smoker<2], label="кластеры")
plt.legend()
plt.show()
```
